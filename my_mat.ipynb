*HISTOGRAM*

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_stu_age=np.random.randint(18,45,(100))
py_stu_age=np.random.randint(15,440,(100))
```

```python
print(ml_stu_age)
print(py_stu_age)
```

```python
bin=[15,20,25,30,45,50]
plt.hist(ml_stu_age,bin,rwidth=.8,histtype="step")
plt.hist(py_stu_age, bin,rwidth=.8, histtype="step")
plt.title("ML Students Age")
plt.xlabel("Students age")
plt.ylabel("No of students")
plt.show()
```
